不同的线条

字符|类型 | 字符|类型
---|--- | --- | ---
`  '-'	`| 实线 | `'--'`|	虚线
`'-.'`|	虚点线 | `':'`|	点线
`'.'`|	点 | `','`| 像素点
`'o'`	|圆点 | `'v'`|	下三角点
`'^'`|	上三角点 | `'<'`|	左三角点
`'>'`|	右三角点 | `'1'`|	下三叉点
`'2'`|	上三叉点 | `'3'`|	左三叉点
`'4'`|	右三叉点 | `'s'`|	正方点
`'p'`	| 五角点 | `'*'`|	星形点
`'h'`|	六边形点1 | `'H'`|	六边形点2 
`'+'`|	加号点 | `'x'`|	乘号点
`'D'`|	实心菱形点 | `'d'`|	瘦菱形点 
`'_'`|	横线点 | |

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'-.')
plt.xlabel('xlabel',fontsize = 16)#fontsize 字体大小
plt.ylabel('ylabel',fontsize = 16)
```

绘制多个线

```python
#一个图里加多条线-可以写在一个plt.plot里，也可以写成多条plt.plot
```

```python
tang_numpy = np.arange(0,10,0.5)
tang_numpy
```

```python
tang_numpy = np.arange(0,10,0.5)
plt.plot(tang_numpy,tang_numpy,'r--')
plt.plot(tang_numpy,tang_numpy**2,'bs')
plt.plot(tang_numpy,tang_numpy**3,'go')
```

```python
plt.plot(tang_numpy,tang_numpy,'r--',
        tang_numpy,tang_numpy**2,'bs',
        tang_numpy,tang_numpy**3,'go')
```

指定线条的宽度

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y,linewidth = 3.0)
```

```python
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize =10)
```

```python
plt.plot(x,y,color='r',linewidth = 2.0, alpha = 0.4)#alpha 是透明度
```

子图

```python
# 211 表示一会要画的图是2行一列的 最后一个1表示的是子图当中的第1个图
plt.subplot(211)
plt.plot(x,y,color='r')

# 212 表示一会要画的图是2行一列的 最后一个1表示的是子图当中的第2个图
plt.subplot(212)
plt.plot(x,y,color='b') 
```

```python
# 211 表示一会要画的图是1行2列的 最后一个1表示的是子图当中的第1个图
plt.subplot(121)
plt.plot(x,y,color='r')

# 212 表示一会要画的图是1行2列的 最后一个1表示的是子图当中的第2个图
plt.subplot(122)
plt.plot(x,y,color='b')
```

```python
plt.subplot(222)
plt.plot(x,y,color='r')


plt.subplot(223)
plt.plot(x,y,color='b')
```

给图上加上注释

```python
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
plt.xlabel('x:---')
plt.ylabel('y:---')
plt.title('cheng qian:---')
plt.text(0,0,'zhangshengnan')
plt.grid(True)
plt.annotate('zhangshengnan',xy=(-5,0),xytext=(-2,0.3),arrowprops = dict(facecolor='red',shrink=0.05,headlength= 20,headwidth = 20))
```
