# 二、Pandas数据操作

* Series索引

```python
ser_obj = pd.Series(range(5), index = ['a', 'b', 'c', 'd', 'e'])
print(ser_obj.head())
```

```python
# 行索引
print(ser_obj['a'])
print(ser_obj[0])
```

```python
# 切片索引
print(ser_obj[1:4])
print(ser_obj['b':'d'])
```

```python
ser_obj
```

```python
# 不连续索引
print(ser_obj[[0, 2, 4]])
print(ser_obj[['a', 'e']])
```

```python
ser_obj>2
```

```python
# 布尔索引
ser_bool = ser_obj > 2
print(ser_bool)
print(ser_obj[ser_bool])

print(ser_obj[ser_obj > 2])
```

* DataFrame索引

```python
import numpy as np

df_obj = pd.DataFrame(np.random.randn(5,4), columns = ['a', 'b', 'c', 'd'])
print(df_obj.head())
```

```python
# 标签索引 loc

# DataFrame
print(df_obj['a'])
print(df_obj.loc[0:2, 'a'])
```

```python
# 整型位置索引 iloc


# DataFrame
print(df_obj.iloc[0:2, 0]) # 注意和df_obj.loc[0:2, 'a']的区别
```

```python
# 混合索引 ix
print(ser_obj.ix[1:3])
print(ser_obj.ix['b':'c'])

# DataFrame
print(df_obj.ix[0:2, 0]) # 先按标签索引尝试操作，然后再按位置索引尝试操作
```
