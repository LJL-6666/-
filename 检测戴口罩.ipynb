```python
import paddlehub as hub
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import warnings
warnings.filterwarnings("ignore")
```

```python
#查看待检测的图片
img =mpimg.imread("test1.jpeg")
plt.imshow(img) # 显示图片
plt.axis('off') # 不显示坐标轴
plt.show()
```

```python
#加载待检测的图片路径
img_path = ["test1.jpeg"]
#载入口罩预测模型
module = hub.Module(name="pyramidbox_lite_mobile_mask")
#输入待预测图片
input_dict = {"image": img_path}
datas = module.face_detection(data=input_dict)
#输出检测结果
for data in datas:
    print(data)
```

```python
#查看检测后的图片
img =mpimg.imread("detection_result/"+str(img_path[0]))
plt.imshow(img) # 显示图片
plt.axis('off') # 不显示坐标轴
plt.show()
```
